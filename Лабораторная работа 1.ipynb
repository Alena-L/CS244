### Работа с wikipedia-API

```python
import wikipediaapi as wikiapi
import re
import json
```

```python
wiki = wikiapi.Wikipedia("ru")
```

```python
mysteries = wiki.page("Категория:Тайны")
print(mysteries.categorymembers.keys(), sep = '\n')
```

```python
pages = list(mysteries.categorymembers.keys())
cats = []
```

```python
for page in pages:
    if re.search(r'Категория', page):
        cats.append(page)

for c in cats:
    pages.remove(c)        
```

```python
cats
```

```python
def divide(categories):
    a = []
    b = []
    
    for c in categories:
        cats_pages = wiki.page(c)
        cp = list(cats_pages.categorymembers.keys())
        
        for i in range(len(cp)):
            if re.search(r'Категория', cp[i]):
                a.append(cp[i])
            else:
                b.append(cp[i])
    return a,b
```

```python
cats_new, pages_new = divide(cats)
```

```python
cats_new
```

```python
pages += pages_new
```

### Работа с JSON

```python
textcount = 0
```

```python
for title in pages:
    
    textcount += 1    
    page = wiki.page(title)
    text = page.summary
    
    count = len(re.findall(r'\w+', text)) # кол-во слов в текстe
    
    doc = {
        'title': title,
        'text': text,
        'wordcount': count
    }
    
    name = 'document' + str(textcount)
    with open(name, 'w', encoding='utf8') as write_f:
        json.dump(doc, write_f, ensure_ascii=False)
```

```python
wordtotal = 0 
counter = 0
```

```python
for i in range(1, textcount+1):
    name = 'document' + str(textcount)
    
    with open(name, "r") as read_file:
        data = json.load(read_file)
    
    counter += 1
    wordtotal += data['wordcount']
```
